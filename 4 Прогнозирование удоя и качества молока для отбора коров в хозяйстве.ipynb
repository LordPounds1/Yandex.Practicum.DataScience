# Описание проекта

Проект направлен на создание моделей для отбора коров в хозяйстве «Вольный луг», которое планирует расширение стада. Фермер заключил контракт с ассоциацией пастбищ «ЭкоФерма», и с их помощью он хочет принимать более обоснованные решения о покупке коров.

Основной задачей проекта является анализ данных о текущем стаде и потенциальных покупках, предоставленных «ЭкоФермой», для разработки двух моделей машинного обучения:

* Прогнозирование годового удоя коровы — модель поможет определить, какая корова может дать не менее 6000 кг молока в год.


* Расчёт вероятности получения вкусного молока — модель оценит вероятность того, что молоко будет соответствовать вкусовым критериям фермера.

### Создание новых категорий в данных

#### Main

Для улучшения анализа и увеличения информативности данных можно объединить некоторые признаки, которые логически связаны друг с другом. Это поможет лучше интерпретировать результаты и повысить предсказательную силу моделей:

* Питательные свойства корма (**`Nutrition_Score`**) - Объеденим 3 признака которые описывают качество корма. Наш новый признак  учитывает общую энергетическую ценность и содержание протеинов в корме.


* Показатели молока (**`Milk_Quality_Score`**) - Объеденим 2 показателя качества молока с точки зрения содержания жиров и белков.  

```python
main['nutrition_score'] = (main['energy_feed'] * 0.4 + 
                           main['raw_protein'] * 0.3 + 
                           main['sugar_protein'] * 0.3)

main['milk_quality_score'] = (main['percentage_of_fat'] * 0.5 + 
                              main['percentage_of_proteins'] * 0.5)
```

Также для дальнейшего анализа, мы создадим 2 новые категории:

1. **`Nutrition_Category`** - разделим значения на несколько категорий, используя квартильное распределение:
    
    * `Низкое качество корма` - `Nutrition_Score` = ниже 25% (537.579) 
    
    * `Среднее качество корма` - `Nutrition_Score` = от 25% до 75% (от 537.579 до 624.199)
    
    * `Высокое качество корма` - `Nutrition_Score` = выше 75% (624.199)
    
2. **`Milk_Quality_Category`**  - разделим значения на несколько категорий, используя квартильное распределение:

    * `Низкое качество молока` - `Milk_Quality_Score` = ниже 25% (3.333)
    
    * `Среднее качество корма` - `Milk_Quality_Score` = от 25% до 75% (от 3.333 до 3.3975)
    
    * `Высокое качество корма` - `Milk_Quality_Score` = выше 75% (3.3975)
    
Nutrition_Category и Milk_Quality_Category помогают упростить анализ данных, делая его более наглядным и интерпретируемым. 

```python
# Определение категорий для Nutrition_Score
def categorize_nutrition(score):
    if score < 537.579:
        return 'Низкое качество корма'
    elif 537.579 <= score <= 624.199:
        return 'Среднее качество корма'
    else:
        return 'Высокое качество корма'

main['nutrition_category'] = main['nutrition_score'].apply(categorize_nutrition)
```

```python
# Определение категорий для Milk_Quality_Score
def categorize_milk_quality(score):
    if score < 3.333:
        return 'Низкое качество молока'
    elif 3.333 <= score <= 3.3975:
        return 'Среднее качество молока'
    else:
        return 'Высокое качество молока'

main['milk_quality_category'] = main['milk_quality_score'].apply(categorize_milk_quality)
```

```python
RANDOM_STATE = 42
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, random_state=RANDOM_STATE)
```

Кодирование категориальных признаков:

Для обработки категориальных признаков мы применим `OneHotEncoder` с параметром `drop='first'`, чтобы избежать проблемы мультиколлинеарности. 

```python
encoder_log = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
categorical_features_log = ['breed', 'type_of_pasture', 'daddys_breed', 'age', 'sugar_protein_binary']
X_train_cat_log = encoder_log.fit_transform(X_train_log[categorical_features_log])
X_test_cat_log = encoder_log.transform(X_test_log[categorical_features_log])
```

Количественные признаки нормализуем с помощью `StandardScaler`

```python
scaler_log = StandardScaler()
numeric_features_log = ['energy_feed_squared', 'raw_protein', 'percentage_of_fat', 'percentage_of_proteins']
X_train_num_log = scaler_log.fit_transform(X_train_log[numeric_features_log])
X_test_num_log = scaler_log.transform(X_test_log[numeric_features_log])
```

Создаем экземпляр модели и обучим ее на подготовленных данных.

```python
X_train_prepared_log = np.hstack([X_train_cat_log, X_train_num_log])
X_test_prepared_log = np.hstack([X_test_cat_log, X_test_num_log])
```

### Анализ модели

Улучшим модель, путем подбора порогов классификации, чтобы минимизировать количество ложноположительных классификаций и повысить Precision.

Для минимизации ошибок нужно сдвинуть порог классификации выше по шкале предсказанных вероятностей. Таким образом, модель будет классифицировать коров как "вкусные" только при более высоких вероятностях.

Предположим, `y_proba` — это предсказанные вероятности, а `y_test` — истинные метки классов

```python
y_proba = model_lr.predict_proba(X_test_prepared_log)[:, 1]
data = pd.DataFrame(zip(y_test_log, y_proba),
columns = ['y_valid', 'y_proba']).sort_values(by='y_proba',ascending=False) 
data
```

Создадим цикл для подбора оптимального порога

```python
thresholds = np.arange(0.0, 1.01, 0.01)

optimal_thresholds = []
for threshold in thresholds:
    y_pred = (data['y_proba'] >= threshold).astype(int)
    
    tn, fp, fn, tp = confusion_matrix(data['y_valid'], y_pred).ravel()
    
    if fp == 0:
        optimal_thresholds.append((threshold, tn, fp, fn, tp))

optimal_thresholds_df = pd.DataFrame(optimal_thresholds, columns=['Threshold', 'TN', 'FP', 'FN', 'TP'])
optimal_thresholds_df
```

**0.81** является для нас оптимальным

```python
threshold = 0.81
y_pred_081 = (y_proba >= threshold).astype(int)
```

Построим матрицу ошибок:

```python
cm_1 = confusion_matrix(y_test_log, y_pred_081)
plt.figure(figsize=(6,4))
sns.heatmap(cm_1, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Предсказанные значения')
plt.ylabel('Истинные значения')
plt.title('Матрица ошибок')
plt.show()
```

```python
precision_new = precision_score(y_test_log, y_pred_081)
recall_new = recall_score(y_test_log, y_pred_081)

print(f"Новый порог: {threshold}")
print(f"Precision: {precision_new:.3f}")
print(f"Recall: {recall_new:.3f}")
```

При пороге 0.85 модель минимизирует ошибки FP — так модель сэкономит больше всего средств заказчику и предотвратит неправильную покупку коровы

### Важные метрики моделей классификации:

* Точность (Accuracy):
    
    * Процент правильно классифицированных экземпляров, однако не всегда надежен при несбалансированных данных.

* Полнота (Recall):

    * Доля истинно положительных результатов от всех положительных классов; критически важна в задачах, где важно не пропустить положительные случаи (например, выявление болезней).
    
* Точность (Precision):

    * Доля истинно положительных результатов от всех предсказанных положительных классов; важна для снижения ложных срабатываний.
    
   
* Коэффициент детерминации (R2):

    * Измеряет долю вариации зависимой переменной, объясненную независимыми переменными в модели. Он принимает значения от 0 до 1, где 1 означает, что модель идеально объясняет все вариации в данных.
    

* Средняя абсолютная ошибка (MAE):

    * MAE измеряет среднюю величину ошибок в прогнозах модели без учета направления. Она рассчитывается как среднее значение абсолютных ошибок.
    
* Среднеквадратическая ошибка (MSE):


    *  MSE измеряет среднюю квадратуру ошибок между предсказанными и фактическими значениями.
    
    
* Корень из среднеквадратической ошибки (RMSE):


    * RMSE является квадратным корнем из MSE и возвращает ошибку в тех же единицах, что и целевая переменная.
